## now implement Stop words and stemming

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

### Dataset Preprocessing

```python

from nltk.stem.porter import PorterStemmer ##stemming purpose
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['title'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
corpus[1]
```

### Onehot Representation

```python
onehot_repr=[one_hot (words,voc_size)for words in corpus]
onehot_repr
```

```python
corpus[1]
```

```python
onehot_repr[1]
```

## Embedding Representation

### all the sentence has diffrent length so we have to make equal size

### so the labirary w r going to use padding

```python
sent_length=20
embedded_docs=pad_sequences(onehot_repr,padding='post',maxlen=sent_length)
print(embedded_docs)


```

```python
embedded_docs[0]
```

```python
voc_size
```

## Creating model

```python

embedding_vector_features=40 ##features representation (every index will be represented by 40 feture with some diffrent diffrent values )
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(LSTM(100))    # neuron 
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```
