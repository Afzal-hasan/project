## Training Word2Vec From Scratch And AvgWord2vec Indepth Inutuition

```python
# importing the Dataset

import pandas as pd

messages = pd.read_csv('SMSSpamCollection.txt', sep='\t',names=["label", "message"])
```

### Data cleaning and preprocessing
####  Tokenization , Stopwords , stamming , lemmatization , NLTK

```python
import re
import nltk
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]

for i in range(0,len(messages)):
    review=re.sub("[^a-zA-Z0-9]"," ",messages["message"][i])
    review=review.lower()
    review=review.split()
    
    review=[ps.stem(word)for word in review if word not in stopwords.words("english")]
    
    review=" ".join(review)
    corpus.append(review)





```

```python
corpus
```

# Creating the Bag of Words model

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2500,binary=True,ngram_range=(2,2))   #max_feture say we can take top 2500 feture ie.. if the word come two time it will no be 1,2, that will be only 1

X=cv.fit_transform(corpus).toarray()
```

## Dependent Variables and independent

```python
y=pd.get_dummies(messages["label"])
y=y.iloc[:,1].values
```

## now split the data for training and test

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=20)
```

## now train the model

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(X_train, y_train)
```

## prediction

```python
y_pred=spam_detect_model.predict(X_test)
```

##  SCORE

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    review = review.lower()
    review = review.split()
    
    review = [lemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
words=[]


for sent in corpus:
    sent_token=sent_tokenize(sent)
    for sent in sent_token:
        words.append(simple_preprocess(sent))
    






```
